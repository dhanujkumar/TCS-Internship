# Outlier Detection & Handling

```python
## Finding outliers
graph=data.select_dtypes(include=['float64'])
graph.boxplot(figsize=(10,5))
plt.show()
```

```python
Q1 = np.percentile(data['AGE'], 25, interpolation = 'midpoint')  
Q2 = np.percentile(data['AGE'], 50, interpolation = 'midpoint')  
Q3 = np.percentile(data['AGE'], 75, interpolation = 'midpoint') 
IQR = Q3 - Q1 
low_lim = Q1 - 1.5 * IQR 
up_lim = Q3 + 1.5 * IQR 
```

```python
print(low_lim)
print(up_lim)
```

```python
outlier =[] 
for x in data['AGE']: 
    if ((x> up_lim) or (x<low_lim)): 
         outlier.append(x)
print(outlier)
```

# Encoding

```python
## Label Encoding
df=data.copy()
```

```python
## Importing the Library for Label Encoder
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()

# Remove the non-breaking space character from the column names
col_names = ["GENDER", "DRUG NAME", "RACE", "SIDE EFFECTS"]
for col_name in col_names:
    col_name = col_name.replace("\u00A0", " ")

# Encode the data
for col in col_names:
    df[col] = label.fit_transform(data[col])

# Define the mapping variable
mapping = label.classes_

# Print the mapping
print(f"{mapping}")
```

```python
df.head()
```

```python
attr = []
for i in df.columns:
    if df[i].dtype == 'object':
        attr.append(i)
labelencoder = LabelEncoder()
mapping =[]
for i in attr:
    df[i] = labelencoder.fit_transform(df[i])
    lemap = dict(zip(labelencoder.classes_,labelencoder.transform(labelencoder.classes_)))
    mapping.append(lemap)
print(f"{mapping}")
```

```python
## Plotting a correlation heatmap 
corrmatrix = df.corr()
plt.subplots(figsize=(14,8))
sns.heatmap(corrmatrix,vmin=-0.4 , vmax=0.9, annot=True, cmap='YlGnBu' , linewidth=0.2)
```
